## Finding Minimum value 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(11)

Y = (X-5)**2

print(X,Y)
```

**Goal** Given  a function f(x), we want to find the value of x that minimizes f.

## visualisation

```python
plt.style.use('seaborn')
plt.plot(X,Y)
plt.ylabel('F(x)')
plt.xlabel('X')
plt.show()
```

```python
x = 10  # selecting random value
lr = 0.1   # Learning Rate
error = []
# 50 step to down hill direction
plt.plot(X,Y)
for i in range(50):
    grad =  2*(x-5)
    x = x - lr*grad
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)
    
```

```python
# plot the value of error
plt.plot(error)
plt.show()
```
